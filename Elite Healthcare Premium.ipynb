# Elite-Healthcare-Premium

### Elite-Healthcare-Premium aims to generate comprehensive data for healthcare insurance companies, facilitating the introduction of premium healthcare plans specifically designed for the active 50+ demographic. This project proposes leveraging specific biometrics to assess eligibility criteria for premium healthcare plans, aligning with rates offered to individuals aged 35 and above. The healthcare product is tailored to meet the health needs of qualified and active individuals aged 50 and above who satisfy the specified criteria.

### Benefits for Healthcare Providers:

    1) Reduced Doctor Visits: Elite-Healthcare-Premium leads to a decrease in doctor visits through its focus     on preventive care and proactive health management.
    2) Healthier 50+ Demographic: The project contributes to the creation of a healthier 50+ group by              encouraging and supporting individuals in maintaining an active and healthy lifestyle.
    3) Long-Term Investment Strategy: By targeting preventive care, the project promotes a long-term investment   strategy, reducing the need for immediate cash liquidity in addressing health issues.
    4) Government Incentives and Tax Credits: Participation in Elite-Healthcare-Premium opens avenues            for healthcare providers to access government incentives and tax credits, fostering financial benefits for      their operations.  
    
### Benefits for 50+ Individuals:

    1) Affordable Healthcare Premiums: Elite-Healthcare-Premium ensures that individuals aged 50 and above enjoy healthcare premiums at the same cost as those for individuals aged 35 and above, making quality healthcare      more accessible.
    2) Healthier Lifestyle: The project encourages and supports a healthier lifestyle for individuals in the 50+ demographic, fostering overall well-being and preventive health measures.
    3) Reinvestment in Health Saving Plans: Participants have the opportunity to reinvest in health savings       plans, providing a strategic approach to long-term health management and financial planning.
    4) Access to Government Incentives and Tax Credits: Enrolling in Elite-Healthcare-Premium allows individuals aged 50+ to access government incentives and potential tax credits, providing additional financial           benefits."

### In addressing the core challenges of this project, our focus will revolve around:

    1) Data Sufficiency: Can we gather a robust dataset to accurately reflect the current health status of        individuals aged 50 and above?
    2) Feature Identification: Are we able to pinpoint the essential features that determine eligibility for a      premium healthcare plan tailored for the 35+ age group?
    3) Optimal Feature Selection: Can we strike the right balance in selecting features that precisely qualify     individuals aged 50 and above for a premium plan aligned with the rates offered to those aged 35 and above?

By addressing these key challenges, we aim to lay the foundation for a successful implementation of the Elite-Healthcare-Premium project.
     
### For this project, I faced the challenge of obtaining the necessary dataset, and I explored two viable options:

    1) Mock Data Creation: Given the privacy concerns associated with the required data, I opted to generate my own mock data. This approach allows me to address privacy issues while still presenting a comprehensive           proposal to fulfill the project's requirements.
    2) Biometric Data Acquisition: Alternatively, I considered sourcing specific biometric data to support the    project. However, due to privacy constraints, I found it more feasible to create mock data internally.

In light of the privacy issues, I have chosen the first option, creating mock data, as the most suitable means to share my proposal and effectively address the business needs of this project.

```python
# Creating Health Data Profile dataset and displaying the first 5 observations

# Set seed for reproducibility
np.random.seed(55)

# Generate synthetic health data
fake = Faker()

# Number of samples in the dataset
num_samples = 1000

# Create a list of random dates within a specific range
start_date = datetime(2023, 1, 1)
end_date = datetime(2024, 1, 1)
date_list = [fake.date_between_dates(start_date, end_date) for _ in range(num_samples)]

# Generate individuals with name, city, state, and birth date 
name = [fake.name() for _ in range(num_samples)]
city = [fake.city() for _ in range(num_samples)]
state = [fake.state() for _ in range(num_samples)]
birthday = [fake.date_of_birth() for _ in range(num_samples)]

# Generate random age between 35 and 65
age_list = [random.randint(35, 65) for _ in range(num_samples)]

# Generate random values for health metrics
body_height = np.round(np.random.normal(loc=170, scale=10, size=num_samples), 2)
body_weight = 1 * body_height + np.round(np.random.normal(loc=0, scale=10, size=num_samples), 2)
heart_rate_list = np.round(np.random.normal(loc=68, scale=10, size=num_samples), 2)
heart_rate_resting = np.round(np.random.normal(loc=65, scale=10, size=num_samples), 2)
heart_rate_active = np.round(np.random.normal(loc=140, scale=20, size=num_samples), 2)
heart_rate_active_veriability = np.round(np.random.normal(loc=88, scale=10, size=num_samples), 2)
blood_pressure_systolic = np.random.randint(90, 120, size=num_samples)
blood_pressure_diastolic = np.random.randint(60, 80, size=num_samples)
body_temperature_list = np.round(np.random.normal(loc=98.6, scale=0.5, size=num_samples), 2)
steps_list = np.random.randint(0, 12000, size=num_samples)
walking_heart_rate = np.round(np.random.normal(loc=75, scale=5, size=num_samples), 2)
sleep_duration_list = np.round(np.random.uniform(5, 9, size=num_samples) )
calories_burned = np.round(np.random.uniform(1200, 2000, size=num_samples), 2)
exercise_time = np.round(np.random.uniform(30, 120, size=num_samples) )
stand_time = np.round(np.random.uniform(10, 12, size=num_samples) )
base_calories_burned = np.round(np.random.uniform(800, 1000, size=num_samples), 2)



# Create a Pandas DataFrame
health_data_profile = pd.DataFrame({
    'name': name,
    'city': city,
    'state': state,
    'birthday': birthday,
    'age': age_list,
    'data_date': date_list,
    'body_height': body_height,
    'body_weight': body_weight,
    'heart_rate': heart_rate_list,
    'heart_rate_resting': heart_rate_resting,
    'heart_rate_active': heart_rate_active,
    'heart_rate_active_veriability': heart_rate_resting_veriability,
    'blood_pressure_systolic': blood_pressure_systolic,
    'blood_pressure_diastolic': blood_pressure_diastolic,
    'body_temperature': body_temperature_list,
    'steps': steps_list,
    'walking_heart_rate': walking_heart_rate,
    'sleep_duration': sleep_duration_list,
    'calories_burned': calories_burned,
    'base_calories_burned': base_calories_burned,
    'stand_time': stand_time,
})

# Save the dataset to a CSV file
health_data_profile.to_csv('mock_health_dataset_profile.csv', index=False)
df = health_data_profile
df.head()
```

#### Tuning Hyperparameters

Model parameters are properties that define individual models.

    e.g. regression coefficients
    e.g. decision tree split locations
    They can be learned directly from the training data.
    
For example, individual Lasso regression models are defined by coefficients for each feature: β0,β1,β2,…, etc. These coefficients can be learned directly from the data, so they are model parameters.

Hyperparameters express "higher-level" structural information about modeling algorithms.

    e.g. strength of the penalty used in regularized regression
    e.g. the number of trees to include in a random forest
    They are set before training the model because they cannot be learned directly from the training data.
    
The key distinction is that model parameters can be learned directly from the training data while hyperparameters cannot.

For example, Lasso regression also has a regularization term that penalizes complexity. The "strength" of that penalty, λ (or alpha in sklearn), can't be learned directly from the data because the penalty actually influences the way the algorithm learns. You must set the penalty strength before training the model, so it's a hyperparameter.

For Lasso and Ridge regularized regression, the most impactful hyperparameter is the strength of the penalty.

In scikit-learn, this is the alpha parameter.
alpha is a positive value, typically between 0 and 10.

When tuning a model pipeline, you must preprend the algorithm name and two underscores

    e.g. 'lasso\__alpha' and 'ridge\__alpha'.

```python
#  Lasso hyperparameters
lasso_hyperparameters = {
    'lasso__alpha' : [0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 1, 5, 10]
}

# Ridge_hyperparameters
ridge_hyperparameters = {
    'ridge__alpha' : [0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 1, 5, 10]
}
```
